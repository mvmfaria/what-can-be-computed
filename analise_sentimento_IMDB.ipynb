```python
import os
import tarfile
import urllib.request
from transformers import pipeline
```

```python
url = "https://ai.stanford.edu/~amaas/data/sentiment/aclImdb_v1.tar.gz"
filename = "aclImdb_v1.tar.gz"
extract_folder = "aclImdb"
```

```python
if not os.path.exists(filename):
    urllib.request.urlretrieve(url, filename)

if not os.path.exists(extract_folder):
    with tarfile.open(filename, "r:gz") as tar:
        tar.extractall()
```

```python
analisador = pipeline("sentiment-analysis", truncation=True)
```

```python
def analisar_amostras(caminho_pasta, label_real, quantidade=3):
    arquivos = os.listdir(caminho_pasta)[:quantidade]

    for arquivo in arquivos:
        caminho_completo = os.path.join(caminho_pasta, arquivo)

        with open(caminho_completo, 'r', encoding='utf-8') as f:
            texto_review = f.read()

            resultado = analisador(texto_review)

            sentimento_ia = resultado[0]['label']
            certeza_ia = resultado[0]['score'] * 100

            print(f"comentario analisado: '{texto_review[:100]}...'")
            print(f"gabarito: {label_real.upper()} | predito: {sentimento_ia} (confiança: {certeza_ia:.2f}%)\n")
```

```python
analisar_amostras("aclImdb/test/pos", "Positivo")
analisar_amostras("aclImdb/test/neg", "Negativo")
```
